```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
```

```python
index_number_titanic=pd.read_csv("index_number_titanic.csv")
```

```python
index_number_titanic.head()
```

```python
index_number_titanic.isnull().sum()
```

```python
df=index_number_titanic[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
```

```python
df
```

```python
x=df.drop("Survived",axis=1)
y=df[["Survived"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
trf1=ColumnTransformer([("age_imp",SimpleImputer(),[2]),
                        ("embarked_imp",SimpleImputer(strategy="most_frequent"),[6])],remainder="passthrough")
```

```python
trf2=ColumnTransformer([("ohe_sex_embarked",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1,6])],remainder="passthrough")
```

```python
pipe=Pipeline([("trf1",trf1),
               ("trf2",trf2)])
```

```python
pipe.fit_transform(x_train)
```
