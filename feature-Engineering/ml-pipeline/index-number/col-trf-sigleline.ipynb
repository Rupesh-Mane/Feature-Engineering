```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline

```

```python
#LOAD DATASET
```

```python
index_number_titanic=pd.read_csv("index_number_titanic.csv")
```

```python
index_number_titanic
```

```python
index_number_titanic.shape
```

```python
index_number_titanic.isnull().sum()
```

```python
df=index_number_titanic[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
```

```python
df
```

```python
#filled missing value in age and embarked column
#age-->mean,embarked-->most_frequent
```

```python
x=df.drop("Survived",axis=1)
y=df[["Survived"]]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train
```

```python
y_train
```

```python
x_train.head(3)
```

```python
#CASE-1
```

```python
trf=ColumnTransformer([("age_imp",SimpleImputer(),[2]),
                       ("embarked_imp",SimpleImputer(strategy="most_frequent"),[6]),
                       ("ohe_sex_embarked",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1,6])],remainder="passthrough")
```

```python
pipe=Pipeline([("trf",trf)])
```

```python
pipe.fit_transform(x_train)
```

```python
pipe.get_feature_names_out()
```

```python
#CASE-2
```

```python
df1=index_number_titanic[["Survived","Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]]
```

```python
df1
```

```python
x=df1.drop("Survived",axis=1)
y=df1[["Survived"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
embarked_pipe=Pipeline([("emb_imp",SimpleImputer(strategy="most_frequent")),
                        ("emb_ohe",OneHotEncoder(sparse_output=False,handle_unknown="ignore"))])
```

```python
trf=ColumnTransformer([("age_imp",SimpleImputer(),[2]),
                       ("ohe_sex",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1]),
                       ("ohe_imp_embarked",embarked_pipe,[6])],remainder="passthrough")
```

```python
pipe=Pipeline([("trf",trf)])
```

```python
pipe.fit_transform(x_train)
```

```python
pipe.get_feature_names_out()
```

```python
pd.DataFrame(pipe.fit_transform(x_train),columns=pipe.get_feature_names_out())
```

```python
pd.DataFrame(pipe.fit_transform(x_train),columns=pipe.get_feature_names_out(),index=x_train.index)
```
