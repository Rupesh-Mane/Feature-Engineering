SINGLE COLUMN TRANSFORMER-->USING COLUMN INDEX NUMBER

```python
trf=ColumnTransformer([("imputer_age",SimpleImputer(),[2]),
                       ("imputer_embarked",SimpleImputer(strategy="most_frequent"),[6]),
                       ("ohe_sex_embarked",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1,6])],remainder="passthrough")
```

```python
pipe=Pipeline([("trf",trf)])
```

```python
trf.get_feature_names_out()
```

```python
pipe_df=pipe.fit_transform(x_train)
```

```python
pipe_df
```

#and also the problem is while using single column transformation-> all transformation run same time ,so if u want to used imputer_embarked,not possible because of all transformer run same time so ohe,used original Encoder.and finally its create embarked column for imputer ,and 0he for embarked_c,embarked_s,embarked_q,and embarked_nan.But its not the write way.
If you want to used imputer Embarked then you have to used separate pipeline.for embarked
like this below is the solution.

```python
x_train.head(2)
```

```python
embarked_pipe=Pipeline([("imputer_embarked",SimpleImputer(strategy="most_frequent")),
                        ("ohe_embarked",OneHotEncoder(sparse_output=False,handle_unknown="ignore"))])
```

```python
trf=ColumnTransformer([("imputer_age",SimpleImputer(),[2]),
                       ("ohe_sex",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),[1]),
                       ("embarked_imp_ohe",embarked_pipe,[6])],remainder="passthrough")
```

```python
trf.get_feature_names_out()
```

```python

pipe=Pipeline([("trf",trf)])
```

```python
df=pipe.fit_transform(x_train)
```

```python
embarked_pipe=Pipeline([("imputer_embarked",SimpleImputer(strategy="most_frequent")),
                        ("ohe_embarked",OneHotEncoder(sparse_output=False,handle_unknown="ignore"))])
```

```python
trf=ColumnTransformer([("imputer_age",SimpleImputer(),["Age"]),
                       ("ohe_sex",OneHotEncoder(sparse_output=False,handle_unknown="ignore"),["Sex"]),
                       ("embarked_imp_ohe",embarked_pipe,["Embarked"])],remainder="passthrough")
```

```python
pipe=Pipeline([("trf",trf)])
```

```python
pipe.fit_transform(x_train)
```
