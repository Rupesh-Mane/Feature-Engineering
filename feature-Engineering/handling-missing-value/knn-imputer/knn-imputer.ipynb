```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.impute import KNNImputer
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("knn_imputer.csv",usecols=["Age","Pclass","Fare","Survived"])
```

```python
df.head()
```

```python
df.isnull().mean()
```

```python
#USING KNN IMPUTER
```

```python
x=df.drop(columns=["Survived"])
y=df["Survived"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
#KNN Imputer
knn=KNNImputer() #--> by default n_neighbors=5 and weights=uniform

x_train_knn=knn.fit_transform(x_train)
x_test_knn=knn.transform(x_test)
```

```python
x_train_knn_df=pd.DataFrame(x_train_knn,columns=x_train.columns)
```

```python
x_train_knn_df.head()
```

```python
x_train_knn_df.isnull().mean()
```

```python
lr=LogisticRegression()    #--> because output column is classification 

lr.fit(x_train_knn,y_train)

y_pred_knn=lr.predict(x_test_knn)

print("LR with KNN Imputer:",accuracy_score(y_test,y_pred_knn))
```

```python
#Using Simple Imputer --> by-default is mean only.

si=SimpleImputer()

x_train_si=si.fit_transform(x_train)
x_test_si=si.transform(x_test)
```

```python
lr2=LogisticRegression()

lr2.fit(x_train_si,y_train)
y_pred_si=lr2.predict(x_test_si)
print("LR with Simple Imputer:",accuracy_score(y_test,y_pred_si))
```

```python
#you can modify knn imputer by changing n_neighbors,weights
```

```python
knn2=KNNImputer(n_neighbors=2)

x_train_knn2=knn2.fit_transform(x_train)
x_test_knn2=knn2.transform(x_test)
```

```python
lr_knn2=LogisticRegression()
lr_knn2.fit(x_train_knn2,y_train)
y_pred_knn2=lr_knn2.predict(x_test_knn2)
print("LR with KNN Imputer (n_neighbors=2):",accuracy_score(y_test,y_pred_knn2))
```

```python
knn2=KNNImputer(n_neighbors=3)

x_train_knn2=knn2.fit_transform(x_train)
x_test_knn2=knn2.transform(x_test)
```

```python
lr_knn2=LogisticRegression()
lr_knn2.fit(x_train_knn2,y_train)
y_pred_knn2=lr_knn2.predict(x_test_knn2)
print("LR with KNN Imputer (n_neighbors=3):",accuracy_score(y_test,y_pred_knn2))
```

```python
knn2=KNNImputer(n_neighbors=10)

x_train_knn2=knn2.fit_transform(x_train)
x_test_knn2=knn2.transform(x_test)
```

```python
lr_knn2=LogisticRegression()
lr_knn2.fit(x_train_knn2,y_train)
y_pred_knn2=lr_knn2.predict(x_test_knn2)
print("LR with KNN Imputer (n_neighbors=10):",accuracy_score(y_test,y_pred_knn2))
```

```python
knn2=KNNImputer(n_neighbors=5,weights="distance")

x_train_knn2=knn2.fit_transform(x_train)
x_test_knn2=knn2.transform(x_test)
```

```python
lr_knn2=LogisticRegression()
lr_knn2.fit(x_train_knn2,y_train)
y_pred_knn2=lr_knn2.predict(x_test_knn2)
print("LR with KNN Imputer (n_neighbors=5,weights='distance'):",accuracy_score(y_test,y_pred_knn2))
```
