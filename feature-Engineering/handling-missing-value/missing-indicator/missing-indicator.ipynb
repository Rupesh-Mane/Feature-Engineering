```python
import pandas as pd
```

```python
#load dataset

df=pd.read_csv("train.csv",usecols=["Age","Fare","Survived"])
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.impute import MissingIndicator
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
```

```python
x=df.drop(columns=["Survived"])
y=df["Survived"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.head()
```

```python
#Without missing indicator
```

```python
si=SimpleImputer()

x_train_si=si.fit_transform(x_train)
x_test_si=si.transform(x_test)
```

```python
lr=LogisticRegression()
lr.fit(x_train_si,y_train)

y_pred=lr.predict(x_test_si)

print(accuracy_score(y_test,y_pred))
```

```python
#with missing indicator
```

```python
mi=MissingIndicator()

x_train_missing=mi.fit_transform(x_train)
x_test_missing=mi.transform(x_test)
```

```python
x_train["Age_NAN"]=x_train_missing
x_test["Age_NAN"]=x_test_missing
```

```python
x_train.sample(5)
```

```python
si2=SimpleImputer()

x_train_trf2=si2.fit_transform(x_train)
x_test_trf2=si2.transform(x_test)
```

```python
lr2=LogisticRegression()

lr2.fit(x_train_trf2,y_train)

y_pred2=lr2.predict(x_test_trf2)

print(accuracy_score(y_test,y_pred2))
```
