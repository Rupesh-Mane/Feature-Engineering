```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("titanic_toy.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=["Survived"])
y=df["Survived"]

    
    
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
si=SimpleImputer(strategy="constant",fill_value=99)
si1=SimpleImputer(strategy="constant",fill_value=999)
```

```python
from sklearn.compose import ColumnTransformer
```

```python
trf=ColumnTransformer([("si",si,["Age"]),
                       ("si1",si1,["Fare"])],remainder="passthrough")
```

```python
x_train_trf=trf.fit_transform(x_train)
x_test_trf=trf.transform(x_test)
```

```python
x_train_trf_df=pd.DataFrame(x_train_trf,columns=trf.get_feature_names_out(),index=x_train.index)
```
