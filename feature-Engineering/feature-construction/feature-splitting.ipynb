```python
import pandas as pd
```

```python
df=pd.read_csv("titanic.csv",usecols=["Name","Sex","Embarked"])
```

```python
df.head(2)
```

```python
df.isnull().sum()
```

```python
df["Embarked"]=df["Embarked"].fillna("C")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
#working on name columns-->split it inti title firstname,lastname
```

```python
#lastname and rest
```

```python
df[["LastName","Rest"]]=df["Name"].str.split(",",expand=True)
```

```python
df.head()
```

```python
df["Title"]=df["Rest"].str.extract(r"([A-Za-z]+)\.")
```

```python
df.head()
```

```python
df["FirstName"]=df["Rest"].str.replace(r"([A-Za-z]+)\.","",regex=True).str.split()
```

```python
df.head()
```

```python
df["FirstName_modify"]=df["Rest"].str.replace(r"([A-Za-z]+)\.","",regex=True).str.replace(","," ")
```

```python
df.head()
```
