```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
```

```python
df=pd.read_csv("titanic.csv",usecols=["Age","Pclass","SibSp","Parch","Survived"])
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df=df.dropna()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=["Survived"])
y=df["Survived"]
```

```python
np.mean((cross_val_score(LogisticRegression(),x,y,scoring="accuracy",cv=20)))
```

```python
import pandas as pd

df = pd.DataFrame({
    "Product": ["Wireless Mouse", "Gaming Keyboard", "Headphones", "Mobile Charger"],
    "Price": [1000, 2000, 1500, 800],
    "Discount": [20, 10, 15, 5]
})



print(df)
```

```python
# Feature Construction: Final Price
    
    
df["FinalPrice"] = df["Price"] - (df["Price"] * df["Discount"] / 100)
```
