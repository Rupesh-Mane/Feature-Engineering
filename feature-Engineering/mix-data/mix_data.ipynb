```python
import pandas as pd
```

```python
df=pd.DataFrame({
    "Room":['A15',"G56","B22","C9"]
})

```

```python
df.to_csv("Room.csv",index=False)
```

```python
df
```

```python
#Extracting alphabet part

df["Room_Letter"]=df["Room"].str.extract('([A-Za-z]+)')

#([A-Za-z]+) means: take alphabet only
```

```python
df
```

```python
#Extracting numerical part

df["Room_Number"]=df["Room"].str.extract('(\d+)') # \d python read escape chr.that why used \\d
```

```python
df["Room_Number"]=df["Room"].str.extract('(\\d+)') # \\d 
```

```python
df
```

```python
df.info()
#here the Room_number column is string need to br int.
```

```python
df["Room_Number"]=df["Room_Number"].astype("int32")
```
