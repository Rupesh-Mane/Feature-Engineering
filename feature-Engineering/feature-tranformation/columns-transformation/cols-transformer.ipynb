WITHOUT COLUMNS_TRANFORMATION

```python
#simpleImputer--->fever column

from sklearn.impute import SimpleImputer
si=SimpleImputer()
x_train_fever=si.fit_transform(x_train[["fever"]])
x_test_fever=si.transform(x_test[["fever"]])
```

```python
#ordinalEncoding--->cough column

oe=OrdinalEncoder(categories=[["Mild","Strong"]])
x_train_cough=oe.fit_transform(x_train[["cough"]])
x_test_cough=oe.transform(x_test[["cough"]])
```

```python
#onehotencoding--->gender,city

ohe=OneHotEncoder(sparse_output=False,drop="first")

x_train_gender_city=ohe.fit_transform(x_train[["gender","city"]])
x_test_gender_city=ohe.transform(x_test[["gender","city"]])
```

WITH COLUMN TRANFORMATION   

```python
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer([
    ("trf1",SimpleImputer(),["fever"]),
    ("trf2",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"]),
    ("trf3",OneHotEncoder(sparse_output=False,drop="first"),["gender","city"]),
    ],remainder="passthrough")
```

```python
ct_train_transformed=ct.fit_transform(x_train)
```

```python
ct_train_transformed
```

```python
ct_train_transformed.shape
```

```python
ct_test_transformed=ct.transform(x_test)
```

```python
ct_test_transformed
```

```python
ct_test_transformed.shape
```

```python
from sklearn.compose import ColumnTransformer

ct=ColumnTransformer([
    ("trf1",SimpleImputer(),["fever"]),
    ("trf2",OrdinalEncoder(categories=[["Mild","Strong"]]),["cough"]),
    ("trf3",OneHotEncoder(sparse_output=False,drop="first",dtype=int),["gender","city"]),
    ],remainder="passthrough")
```

```python
ct_train_transformed=ct.fit_transform(x_train)
ct_test_transformed=ct.transform(x_test)
```
