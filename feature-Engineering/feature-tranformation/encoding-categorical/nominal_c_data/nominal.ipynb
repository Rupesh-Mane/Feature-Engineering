```python
import numpy as np
import pandas as pd
```

```python
car=pd.read_csv("cars.csv")
```

```python
car.head() 
#here brand,fuel,owner is nominal categorical data,however we are using only two nominal data here
```

onehotencoding using pandas

```python
pd.get_dummies(car,columns=["fuel","owner"])
```

```python
pd.get_dummies(car,columns=["fuel","owner"],dtype=int)
```

```python
#to solve problem of multicollinearity drop on -one column from the fuel,owner
pd.get_dummies(car,columns=["fuel","owner"],dtype=int,drop_first=True)
```

onehotencodeing using sklearn

```python
car.head()
```

```python
from sklearn.model_selection import train_test_split

x=car[["brand","km_driven","fuel","owner"]]
y=car["selling_price"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder()

x_train_transform=ohe.fit_transform(x_train[["fuel","owner"]])
x_test_transform=ohe.transform(x_test[["fuel","owner"]])
```

```python
x_train_transform
```

```python
x_test_transform
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse_output=False)#sparse_output-->output will be normal array(easy to see)

x_train_transform=ohe.fit_transform(x_train[["fuel","owner"]])
x_test_transform=ohe.transform(x_test[["fuel","owner"]])
```

```python
x_train_transform
```

```python
x_test_transform
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse_output=False,dtype=int)#sparse_output-->output will be normal array(easy to see)

x_train_transform=ohe.fit_transform(x_train[["fuel","owner"]])
x_test_transform=ohe.transform(x_test[["fuel","owner"]])
```

```python
#how to solve multicollinearity problem,for that used parameter,drop="first"

from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse_output=False,dtype=int,drop="first")#sparse_output-->output will be normal array(easy to see)

x_train_transform=ohe.fit_transform(x_train[["fuel","owner"]])
x_test_transform=ohe.transform(x_test[["fuel","owner"]])
```

```python
x_train_transform
```

```python
x_test_transform
```

```python
np.hstack((x_train[["brand","km_driven"]].values,x_train_transform))
```
