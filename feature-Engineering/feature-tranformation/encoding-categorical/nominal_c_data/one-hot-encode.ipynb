```python
from sklearn.model_selection import train_test_split

x=df[["brand","km_driven","fuel","owner"]]
y=df["selling_price"]

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe=OneHotEncoder(sparse_output=False,dtype=int,drop="first")#sparse_output-->output will look normal array,drop-->multicollinearity problem,dtype-->change dummy_variable dtype.

#transform

x_train_transform=ohe.fit_transform(x_train[["fuel","owner"]])
x_test_transform=ohe.transform(x_test[["fuel","owner"]])
```

```python
x_train_transform
```

```python
x_train_remain_data=x_train[["brand","km_driven"]]
x_test_remain_data=x_test[["brand","km_driven"]]
```

```python
x_train_remain_data 
```

```python
x_test_remain_data
```

```python
# now convert array format into dataframe
 
pd.DataFrame(x_train_transform,columns=ohe.get_feature_names_out(["fuel","owner"])).head(5)
#need to write index=x_train_remain_data,because if we ont use it ,df start from 0,1,2 etc
```

```python
x_train_transform_df=pd.DataFrame(x_train_transform,columns=ohe.get_feature_names_out(["fuel","owner"]),index=x_train_remain_data.index)
x_train_transform_df
```

```python
x_test_transform_df=pd.DataFrame(x_test_transform,columns=ohe.get_feature_names_out(["fuel","owner"]),index=x_test_remain_data.index)
x_test_transform_df
```

```python
#now concat remaining & new transform data

pd.concat([x_train_remain_data,x_train_transform_df],axis=1)
```

```python
pd.concat([x_test_remain_data,x_test_transform_df],axis=1)
```
