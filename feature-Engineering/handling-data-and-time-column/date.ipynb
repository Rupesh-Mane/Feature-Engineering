```python
import pandas as pd
```

```python
date=pd.read_csv("orders.csv")
```

```python
date.head()
```

```python
date.shape
```

```python
date.isnull().sum()
```

```python
date.info()
```

```python
#here,the date column dtype is string -->convert it into datetime dtype ,
#because,if it is str we cannot use datetime functionality on date column.
```

```python
date["date"]=pd.to_datetime(date["date"])
```

```python
date.head()
```

```python
date.info()
#Now date column dtype is datetime,now we can use datetime functionality.
```

```python
#Extract year

date["Year"]=date["date"].dt.year
```

```python
date.head()
```

```python
date=date.drop(columns=["product_id","city_id","orders"])
```

```python
date.head()
```

```python
date=date.rename(columns={"date":"Date"})
```

```python
date.head()
```

```python
#Extract month

date["Month"]=date["Date"].dt.month
```

```python
date.head()
```

```python
#month name
date["Month_Name"]=date["Date"].dt.month_name()
```

```python
date.head()
```

```python
#Extract day

date["Day"]=date["Date"].dt.day
```

```python
date.head()
```

```python
#day name

date["Day_Name"]=date["Date"].dt.day_name()
```

```python
date
```

```python
#Extract date_of_week

date["Date_dow"]=date["Date"].dt.day_of_week
```

```python
date.head()
```

```python
#is_weekend
import numpy as np
date["Date_Weekend"]=np.where(date["Day_Name"].isin(["Saturday","Sunday"]),1,0)
```

```python
date.head()
```

```python
date=date.drop(columns=["Week_Of_The_Year"])
```

```python
date.head()
```

```python
#Extract quarter

date["Quarter"]=date["Date"].dt.quarter
```

```python
date.head()
```

```python
#Extract semester

date["Semester"]=np.where(date["Quarter"].isin([1,2]),1,2)
```
